```python
import pandas as pd
import numpy as np
```

```python
amazondf = pd.read_csv("./archivos/amazon_prime_titles.csv")
disneydf = pd.read_csv("./archivos/disney_plus_titles.csv")
huludf = pd.read_csv("./archivos/hulu_titles.csv")
netflixdf = pd.read_csv("./archivos/netflix_titles.csv")
```

Generar campo id: Cada id se compondrá de la primera letra del nombre de la plataforma, seguido del show_id ya presente en los datasets (ejemplo para títulos de Amazon = as123)

```python
amazondf['id']= amazondf['show_id'].apply(lambda i: 'a'+i.split()[0])
disneydf['id']= disneydf['show_id'].apply(lambda i: 'd'+i.split()[0])
huludf['id']= huludf['show_id'].apply(lambda i: 'h'+i.split()[0])
netflixdf['id']= netflixdf['show_id'].apply(lambda i: 'n'+i.split()[0])
```

Agregar una columna a cada dataframe para saber cada Movie o Tv Show a qué plataforma pertenece.

```python
netflixdf['platform'] = 'Netflix'
huludf['platform'] = 'Hulu'
disneydf['platform'] = 'Disney'
amazondf['platform'] = 'Amazon'
```

```python
df = pd.concat([amazondf, netflixdf, huludf, disneydf], axis=0)
df.shape
```

Los valores nulos del campo rating deberán reemplazarse por el string “G” (corresponde al maturity rating: “general for all audiences”)

```python
df['rating'].fillna('G', inplace=True)
```

De haber fechas, deberán tener el formato AAAA-mm-dd

```python
df['date_added'] = pd.to_datetime(df['date_added'], infer_datetime_format=True)
df.head()
```

El campo duration debe convertirse en dos campos: duration_int y duration_type. El primero será un integer y el segundo un string indicando la unidad de medición de duración: min (minutos) o season (temporadas)

```python
df[[ 'duration_int','duration_type']]=df['duration'].str.split(' ', expand=True) 
```

```python
df.fillna('Sin dato', inplace=True)
df.sample(3) #Vemos 3 ejemplos
```

Los campos de texto deberán estar en minúsculas, sin excepciones

```python
df = df.apply(lambda x: x.astype(str).str.lower())
df.head()
```

Exportamos el dataset para  realizar las consultas con FastAPI.

```python
df.to_csv('películas.csv', encoding='utf-8', index=False)
```
